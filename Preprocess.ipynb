```python
import pandas as pd
from timezonefinder import TimezoneFinder
import country_converter as coco
import os, sys
import json
import time
```

```python
remove_countries = ["USA"]

# Read TICON-4 data
df = pd.read_csv("data/TICON-4.csv", sep=",")
# df = df[~df["country"].isin(remove_countries)]
df["name"] = None
pd.set_option("display.max_rows", None)
```

```python
# Count unique stations by source
# (assuming stations are uniquely identified by lat/lon)
country_by = df.groupby(["lat", "lon", "country"]).size().reset_index()
country_counts = country_by["country"].value_counts()

print("Countries:")
print(country_counts)
print(f"\nTotal unique stations: {len(country_by)}")
print(f"Total countries: {len(country_counts)}")
```

```python
# Count unique stations by source
# (assuming stations are uniquely identified by lat/lon)
gesla_source_by = df.groupby(["lat", "lon", "gesla_source"]).size().reset_index()
gesla_source_counts = gesla_source_by["gesla_source"].value_counts()

print("TICON-4 Stations by Source:")
print(gesla_source_counts)
print(f"\nTotal unique stations: {len(gesla_source_by)}")
print(f"Total sources: {len(gesla_source_counts)}")
```

```python
datum_information = df.groupby(["lat", "lon", "datum_information"]).size().reset_index()
datum_counts = datum_information["datum_information"].value_counts()

# Count datums
print("TICON-4 Datums:")
print(datum_counts)
print(f"\nTotal unique stations: {len(datum_information)}")
print(f"Total datums: {len(datum_counts)}")
```

```python
record_qualty_by = df.groupby(["lat", "lon", "record_quality"]).size().reset_index()
record_qualty_counts = record_qualty_by["record_quality"].value_counts()

# Count datums
print("TICON-4 Record Quality:")
print(record_qualty_counts)
print(f"\nTotal unique stations: {len(record_qualty_by)}")
print(f"Total unique quality: {len(record_qualty_counts)}")
```
